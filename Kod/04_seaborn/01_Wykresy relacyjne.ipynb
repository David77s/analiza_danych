### Import biblioteki seaborn

```python
import seaborn as sns
```

### Scatter

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip')
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip', hue='sex')
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip', hue='sex', size='size')
```

Line

```python
ice = sns.load_dataset('seaice')
ice.head()
```

```python
ice.info()
```

```python
_ = sns.relplot(ice[ice['Date'].dt.year == 2018], x='Date', y='Extent', kind='line')
```

```python
ice['mm-dd'] = ice['Date'].dt.strftime('%m-%d')
```

```python
ice.head()
```

```python
ice_200_2018 = ice[(ice['Date'].dt.year > 1999) & (ice['Date'].dt.year < 2019)]
```

```python
_ = sns.relplot(ice_200_2018, x = 'mm-dd', y='Extent', kind='line')
```
